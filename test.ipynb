```python
import firebase_admin
from firebase_admin import credentials, firestore

# Initialize Firebase Admin SDK
# Option 1: Using service account (download from Firebase Console > Project Settings > Service Accounts)
cred = credentials.Certificate("credentials.json")

# Option 2: Using Application Default Credentials (if running on GCP or with GOOGLE_APPLICATION_CREDENTIALS env var)
# cred = credentials.ApplicationDefault()

firebase_admin.initialize_app(cred, {
    'projectId': 'synckit-example',
})

# Get Firestore instance
db = firestore.client()

print(f"Connected to Firestore project: synckit-example")
print(f"Firestore client: {db}")
```
